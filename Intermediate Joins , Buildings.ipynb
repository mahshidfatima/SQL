```python
q="""
    select il.invoice_id Invoice_ID , t.track_id Track_ID, t.name Track_Name  , il.quantity Quality ,il.unit_price Unit_Price \
    , m.name Media , al.title Album , ar.name Artist
    from invoice_line il
    inner join track t
    on il.track_id = t.track_id
    inner join media_type m
    on m.media_type_id = t.media_type_id
    inner join album al
    on t.album_id = al.album_id
    inner join artist ar
    on al.artist_id = ar.artist_id
    where invoice_id=2
"""
read_query(q)

```

```python
# album title ,artist name, no of purchase , 
q="""
    select al.title Album_Title, ar.name Artist_Name,sum(il.quantity) Sold from album al
    inner join artist ar
    on al.artist_id = ar.artist_id
    inner join track t
    on al.album_id = t.album_id
    inner join invoice_line il
    on il.track_id = t.track_id
    group by al.album_id
    order by 3 DESC
    
"""
read_query(q)
```

```python
q= """
    select c.first_name || " " || c.last_name 'Full Name' , c.city , c.country , sum(iv.total) total, e.first_name from customer c
    inner join invoice iv
    on c.customer_id = iv.customer_id
    inner join employee e
    on c.support_rep_id = e.employee_id
    group by iv.customer_id
    order by total DESC
"""
read_query(q)
```

```python
q="""
    select sum(il.quantity) Sold ,g.genre_id ,count(g.name) Genre , g.name Name from album al
    inner join artist ar
    on al.artist_id = ar.artist_id
    inner join track t
    on al.album_id = t.album_id
    inner join invoice_line il
    on il.track_id = t.track_id
    inner join genre g
    on g.genre_id = t.genre_id
    group by al.album_id
    order by 3 DESC
    
"""
read_query(q)
```

```python
q="""
    select c.first_name || " " || c.last_name 'Full Name',count(i.invoice_id) 'Total invoices', sum(i.total) 'Total' ,\
    case
        when sum(i.total) < 40 then "Small Spender"
        when sum(i.total) >100 then "Big Spender"
        else "Regular"
    end Customer_category
    from customer as c
    left join invoice i
    on c.customer_id = i.customer_id
    group by c.customer_id
    order by 3 dESC
"""

read_query(q)['Customer_category'].value_counts()
```

```python
#playlist_id, playlist_name , no_of_tracks , totla_durationin secods
q= """
    select p.playlist_id, p.name , count(pt.track_id) 'Total Tracks',(sum(t.milliseconds/1000)) 'time' from playlist as p
    left join playlist_track pt
    on p.playlist_id = pt.playlist_id
    inner join track t
    on pt.track_id = t.track_id
    group by p.playlist_id
    order by 4 desc
"""

read_query(q)
```

```python
#playlist_id, playlist_name , no_of_tracks , totla_durationin secods
q= """with playlist_info as 
    (select p.playlist_id, p.name , pt.track_id 'Total_Tracks',t.milliseconds/1000 'time' from playlist as p
    left join playlist_track pt
    on p.playlist_id = pt.playlist_id
    inner join track t
    on pt.track_id = t.track_id)
    
    select playlist_id, name,count(Total_Tracks),sum(time) from playlist_info
    group by 1
    order by 4 desc
    """

read_query(q)
```

```python
q="""create view customer_above_90 as
    select c.* ,sum(i.total) Total from customer as c
    left join invoice i
    on
    c.customer_id = i.customer_id
    group by c.customer_id
    having sum(i.total) > 90
"""
conn.execute(q)
```

```python
q="""
    select * from customer_above_90 where total<100
"""
read_query(q)
```
