```python
q = """
    select
    f.id , c.facts_id,
    f.name Country, c.name City,
    f.population Country_pop, c.population City_pop
    from facts as f 
    inner join 
    cities as c
    on
    f.id == c.facts_id
"""
read(q)
```

```python
q= """
select 
f.name Country , c.name City from facts as f
left join cities as c
on f.id==c.facts_id
"""
read(q)
```

```python
q= """
select 
f.name Country , c.name City from facts as f
left join cities as c
on f.id==c.facts_id
where City is null
"""
read(q)
```

```python
q = """
    select 
    f.name Country , sum(c.population) URBAN,f.population  TOTAL, cast(sum(c.population) as float)/ f.population PERC
    from facts as f
    inner join cities as c
    on f.id==c.facts_id
    group by f.name
    
"""
read(q)
```

```python
q = """
    select facts_id,sum(population) urban_pop from cities
    group by facts_id
"""
read(q) ####>>>>>>a
```

```python
q = """
    select 
    f.name Country , a.urban_pop URBAN,f.population  TOTAL, cast(a.urban_pop as float)/ f.population PERC
    from facts as f
    inner join (select facts_id,sum(population) urban_pop from cities
    group by facts_id) as a
    on f.id==a.facts_id
    group by f.name
    
"""
read(q)
```
